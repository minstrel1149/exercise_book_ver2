```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats as ss
import statsmodels.api as sm
```

```python
bacteria=np.array([175,190,215,198,184,207,210,193,196,180])
```

```python
avg, std, ste = np.mean(bacteria), np.std(bacteria, ddof=1), ss.sem(bacteria)
avg, std, ste
```

```python
t = ss.t.ppf(0.95, len(bacteria) - 1)
[avg - t * ste, avg + t * ste]
```
