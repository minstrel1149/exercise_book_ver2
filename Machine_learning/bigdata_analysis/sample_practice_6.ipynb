```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV, cross_val_score, train_test_split
from sklearn.ensemble import RandomForestClassifier

X_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/shipping/X_train.csv")
X_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/shipping/X_test.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/shipping/y_train.csv")

train_id = X_train.loc[:, ['ID']]
test_id = X_test.loc[:, ['ID']]

X_train = pd.get_dummies(X_train.drop('ID', axis=1))
X_test = pd.get_dummies(X_test.drop('ID', axis=1))
y_train = y_train.loc[:, 'Reached.on.Time_Y.N']

#rf_clf = RandomForestClassifier(n_estimators=100, random_state=0)
#param_grid = {'max_depth':[5, 7, 9], 'min_samples_leaf':[9, 11, 13, 15]}
#grid = GridSearchCV(rf_clf, param_grid=param_grid, n_jobs=-1, cv=5).fit(X_train, y_train)
#grid.best_params_, grid.best_score_

#X_train_tr, X_train_val, y_train_tr, y_train_val = train_test_split(X_train, y_train, random_state=0)

#rf_clf = RandomForestClassifier(n_estimators=100, max_depth=7, min_samples_leaf=13, random_state=0).fit(X_train_tr, y_train_tr)
#rf_clf.score(X_train_val, y_train_val)

rf_clf = RandomForestClassifier(n_estimators=100, max_depth=7, min_samples_leaf=13, random_state=0).fit(X_train, y_train)
y_pred = test_id.assign(**{'Reached.on.Time_Y.N':rf_clf.predict(X_test)})
y_pred.to_csv('data/etc_practice/submission2.csv', index=False)
```
