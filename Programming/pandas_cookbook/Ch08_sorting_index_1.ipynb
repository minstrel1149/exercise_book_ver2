```python
def remove_binary_cols(df):
    binary_only = df.nunique().eq(2)
    cols = binary_only.loc[binary_only].index
    return df.drop(cols, axis=1)

def select_rows_with_max_cols(df):
    max_cols = df.idxmax()
    unique = max_cols.unique()
    return df.loc[unique]
```
