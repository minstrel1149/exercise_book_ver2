```python
emp = pd.read_csv(p1 / 'employee.csv')
dept_sal = (emp.loc[:, ['DEPARTMENT', 'BASE_SALARY']]
            .sort_values(['DEPARTMENT', 'BASE_SALARY'], ascending=[True, False])
)
dept_sal.head()
```

```python
max_dept_sal = dept_sal.drop_duplicates(subset='DEPARTMENT', keep='first')
max_dept_sal.head()
```

```python
max_dept_sal = max_dept_sal.set_index('DEPARTMENT')
emp = emp.set_index('DEPARTMENT')
```

```python
emp = emp.assign(MAX_DEPT_SALARY=max_dept_sal['BASE_SALARY'])
```
