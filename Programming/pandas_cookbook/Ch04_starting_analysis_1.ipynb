```python
movie = pd.read_csv(p1 / 'movie.csv')
cols = ['movie_title', 'imdb_score', 'budget']
mov2 = movie[cols]
```

```python
mov2.nlargest(100, 'imdb_score').nsmallest(10, 'budget')
```

```python
mov2[mov2['imdb_score'] >= mov2['imdb_score'].nlargest(100).min()].nsmallest(5, 'budget')
```

```python
cols2 = ['movie_title', 'title_year', 'imdb_score']
mov3 = movie[cols2].dropna(how='any')
```

```python
mov3.head()
```

```python
mov3.groupby('title_year', as_index=False)['imdb_score'].max()
```

```python
(mov3.sort_values(by=['title_year', 'imdb_score'], ascending=[True, False])
 .drop_duplicates(subset='title_year', keep='first')
)
```

```python
(mov3.groupby('title_year', as_index=False)
 .apply(lambda df: df.sort_values('imdb_score', ascending=False))
)
```

```python
(mov3.groupby('title_year', as_index=False)
 .apply(lambda df: df.sort_values('imdb_score', ascending=False).head(1))
 .droplevel(0, axis=0)
)
```
