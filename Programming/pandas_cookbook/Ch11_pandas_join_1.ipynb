```python
bball_16 = pd.read_csv(p1 / 'baseball16.csv')
bball_16.head()
```

```python
data_dict = bball_16.iloc[0].to_dict()
data_dict
```

```python
new_data = {k: '' if isinstance(v, str) else np.nan for k, v in data_dict.items()}
new_data
```

```python
random_data = []
for i in range(1000):
    d = {k: np.random.choice(list('abcde')) if isinstance(v, str) else np.random.randint(100) for k, v in data_dict.items()}
    random_data.append(pd.Series(d, name=i+len(bball_16)).to_frame().T)
random_data[:10]
```

```python
years = 2016, 2017, 2018
stocks = [pd.read_csv(p1 / f'stocks_{year}.csv', index_col='Symbol') for year in years]
stock16, stock17, stock18 = stocks
stock16
```

```python
pd.concat(stocks, keys=list(years), names=['Year', 'Symbol'])
```

```python
stock16.merge(stock17, left_index=True, right_index=True, how='outer')
```

```python
pd.concat(stocks, keys=years)
```

```python
stock16.add_suffix('_2016')
```

```python
other = [stock17.add_suffix('_2017'), stock18.add_suffix('_2018')]
stock16.add_suffix('_2016').join(other, how='outer')
```

```python
(pd.concat(stocks, keys=years, axis=1)
 .swaplevel(0, 1, axis=1)
 .pipe(lambda df: df.set_axis(df.columns.to_flat_index(), axis=1))
 .rename(lambda label: '_'.join([str(x) for x in label]), axis=1)
)
```

```python
prac = (pd.concat(stocks, keys=years, axis=1)
 .swaplevel(0, 1, axis=1)
)
prac.set_axis(prac.columns.to_flat_index(), axis=1)
```

```python
(stock16
 .merge(stock17, left_index=True, right_index=True, how='outer', suffixes=['_2016', '_2017'])
 .merge(stock18.add_suffix('_2018'), left_index=True, right_index=True, how='outer')
)
```
