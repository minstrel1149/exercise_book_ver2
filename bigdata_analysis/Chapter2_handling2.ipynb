```python
score = pd.DataFrame({'국어': [100, 80], '수학':[75, 90], 
                      '영어':[90, 95]}, index=['장화', '홍련'])
score
```

```python
stud = pd.DataFrame({'국어': [70, 85], '수학':[65, 100], 
                             '영어':[95, 65]}, index=['콩쥐', '팥쥐'])
score = score.append(stud)
score
```

```python
score2 = pd.concat([score, stud], axis=0)
score2 = score2.append(pd.Series({'국어':85, '수학':55, '영어':95}, name='해님'))
score2
```

```python
condition_list = [(score2['국어'] >= 90),
                 (score2['국어'] >= 80) & (score2['국어'] < 90),
                 (score2['국어'] >= 70) & (score2['국어'] < 80)]
choice_list = ['A', 'B', 'C']
```

```python
score2['점수'] = np.select(condition_list, choice_list, default='F')
score2
```

```python
iris = load_iris()
iris_df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_df.head()
```

```python
iris_df['class'] = iris['target']
iris_df['class'] = iris_df['class'].map(dict(zip([0, 1, 2], ['setosa', 'versicolor', 'virginica'])))
iris_df.head()
```

```python
iris_df.groupby(['class']).mean()
```

```python
iris_df['petal with level'] = pd.qcut(iris_df['petal width (cm)'], q=3, labels=['short', 'middle', 'long'])
```

```python
pd.crosstab(iris_df['petal with level'], iris_df['class'], normalize='index')
```
