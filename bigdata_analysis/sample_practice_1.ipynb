```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score

train_df = pd.read_csv('data/car_segment/train.csv')
test_df = pd.read_csv('data/car_segment/test.csv')
X_train = pd.get_dummies(train_df.iloc[:, 1:-1])
y_train = train_df.iloc[:, -1]
X_test = pd.get_dummies(test_df.iloc[:, 1:])

rf_clf = RandomForestClassifier(n_estimators=100, max_depth=10, n_jobs=-1, random_state=0).fit(X_train, y_train)
#cv = cross_val_score(rf_clf, X_train, y_train, scoring='f1_macro', cv=5)

#rf_clf.predict(X_test)
submit_pred = pd.DataFrame({'ID':test_df.loc[:, 'ID'], 'Segmentation':rf_clf.predict(X_test)})
submit_pred.to_csv('data/car_segment/submission_2.csv', index=False)

#X_train_tr, X_train_vd, y_train_tr, y_train_vd = train_test_split(X_train, y_train, random_state=0)
#rf_clf = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=0).fit(X_train_tr, y_train_tr)
#f1_score_macro = f1_score(y_train_vd, rf_clf.predict(X_train_vd), average='macro')
#rf_clf.score(X_train_tr, y_train_tr), rf_clf.score(X_train_vd, y_train_vd), f1_score_macro
```
