```python
score = {'학년':[1, 1, 1, 1, 2, 2], 
         '반':['A', 'A', 'B', 'B', 'C', 'C'], 
         '성별':['여자', '남자', '여자', '남자', '여자', '남자'],
         '성적': [76, 88, 85, 72, 68, 70]}

score = pd.DataFrame(score)
score
```

```python
(score
.set_index(['학년', '반', '성별'])
.unstack()
)
```

```python
score2 = (score
.pivot(index=['학년', '반'], columns='성별', values='성적')
)
score2
```

```python
(score2
.reset_index()
.melt(id_vars=['학년', '반'], var_name='성별', value_name='성적')
)
```
