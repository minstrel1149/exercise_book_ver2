```python
import numpy as np
```

```python
def corr_and_cosine(x, y):
    numerator = np.dot(x, y)
    denominator = np.linalg.norm(x) * np.linalg.norm(y)
    cosine = numerator / denominator

    xm = x - np.mean(x)
    ym = y - np.mean(y)
    numerator_corr = np.dot(xm, ym)
    denominator_corr = np.linalg.norm(xm) * np.linalg.norm(ym)
    correlation = numerator_corr / denominator_corr

    return correlation, cosine
```

```python
a = np.random.randn(15)
b = np.random.randn(15)
a, b
```

```python
cos, corr = corr_and_cosine(a, b)
cos, corr
```
